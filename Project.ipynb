```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("moviedata.csv")
print(df)
```

```python
plt.figure(figsize=(12,5))
plt.bar(df["Movie"], df["Rating"])
plt.xticks(rotation=90)
plt.title("Movie Ratings")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Voting"], df["Rating"])
plt.xlabel("Votes")
plt.ylabel("Rating")
plt.title("Votes vs Rating")
plt.show()
```

```python
print(df.columns)
```

```python
genre_count = df["Genre"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(genre_count, labels=genre_count.index, autopct='%1.1f%%')
plt.title("Genre Distribution")
plt.show()
```

```python
plt.hist(df["Rating"], bins=5)
plt.xlabel("Rating")
plt.title("Rating Distribution")
plt.show()
```

```python
plt.boxplot(df["Rating"])
plt.title("Boxplot of Ratings")
plt.show()
```

```python
plt.barh(df["Movie"], df["Rating"])
plt.xlabel("Rating")
plt.title("Movie Ratings")
plt.show()
```

```python
plt.plot(df["Movie"], df["Rating"], label="Rating", marker='o')
plt.plot(df["Movie"], df["Voting"], label="Votes", marker='s')
plt.xticks(rotation=45)
plt.legend()
plt.title("Rating & Votes Comparison")
plt.show()
```

```python
plt.plot(df["Year"], df["Rating"], marker='o')
plt.xlabel("Year")
plt.ylabel("Rating")
plt.title("Movie Ratings Over Years")
plt.show()
```

```python
plt.fill_between(df["Year"], df["Rating"], alpha=0.5)
plt.title("Rating Area Chart")
plt.xlabel("Year")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.bar(df["Movie"], df["Voting"], color='orange')
plt.xticks(rotation=80)
plt.title("Movie Votes")
plt.ylabel("Votes")
plt.show()
```

```python
plt.scatter(df["Year"], df["Rating"])
plt.title("Rating by Release Year")
plt.xlabel("Year")
plt.ylabel("Rating")
plt.show()
```
